## This is the face mask detector.

The following code utilizes Faster R-CNN in order to perform object detection and detec face masks on people in a picture.

#### Store image's directory into a list

```python
#retrieve image imformation
import os
image_directories = []
for dirname, _, filenames in os.walk('archive/images/'):
    for filename in filenames:
#         print(os.path.join(dirname, filename))
        image_directories.append(os.path.join(dirname, filename))
```

#### Create Dataframe based on image's data found in XML

```python
import xml.etree.ElementTree as et
import re
import pandas as pd
import matplotlib.pyplot as plt

# Create dictionary of dimensions
dic = {"image": [],"Dimensions": []}
for i in range(1,116):
	dic[f'Object {i}']=[]
print("Generating data in CSV format....")

img_elem = [] # The amount of masks in image
img_count = [] # not used

# Parse through image data (xml)
for file in os.listdir("archive/annotations"):
    row = []
    xml = et.parse("archive/annotations/"+file) 
    root = xml.getroot()
    
    img = root[1].text
    row.append(img)
    
    num_img = int(img[12:-4])
    img_count.append(num_img)
    
    h,w = root[2][0].text,root[2][1].text
    count = 0
    row.append([h,w])
    temp_img = []
    for i in range(4,len(root)):
        temp = []
        if root[i][0].text != "mask_weared_incorrect":
            
            temp.append(root[i][0].text)
            count +=1
    #         print(root[i][0].text)
            temp_img.append(root[i][0].text) 
            for point in root[i][5]:
                temp.append(point.text)
            row.append(temp)
#     img_elem.append(temp_img)
    img_elem.append(count)
    
    for i in range(len(row),119):
        row.append(0)
    for i,each in enumerate(dic):
        dic[each].append(row[i])

#Main data
df = pd.DataFrame(dic)
#bounding box information for understanding data
df2 = pd.DataFrame(img_elem)
```

### Data Augmentation

```python
import copy
import random as random
import numpy as np
import matplotlib.pyplot as plt

is_save = False

for i in df.index.values:
    img = cv2.imread('archive/images/' + df["image"][i])

    aug_dict = {}
    
    img_aug_name = df["image"][i][:-4] + "a" + ".png"
    img_aug_dest = 'archive/augmented_data/images/' + img_aug_name
    aug_dict["image"] = img_aug_name
    
    image_directories.append(img_aug_dest)
    
    rows = int(df["Dimensions"][i][0])
    cols = int(df["Dimensions"][i][1])
    aug_dict["Dimensions"] = [rows, cols]

    for j in range(2, 41):
        if df["Object " + str(j - 1)][i] == 0:
            aug_dict["Object " + str(j - 1)] = 0
            continue
            
        # Horizontal Reflection
        bbox = copy.deepcopy(df["Object " + str(j - 1)][i])
        bbox[1] = int(cols) - int(bbox[1])
        bbox[3] = int(cols) - int(bbox[3])
        
        aug_dict["Object " + str(j - 1)] = bbox
    
    df = df.append(aug_dict, ignore_index=True)
    
    img_horiz = cv2.flip(img, 1)
    
    if is_safe:
        plt.imsave(img_aug_dest, img_horiz)


    # Vertical Reflection
    aug_dict = {}
    
    img_aug_name = df["image"][i][:-4] + "b" + ".png"
    img_aug_dest = 'archive/augmented_data/images/' + img_aug_name
    aug_dict["image"] = img_aug_name
    
    aug_dict["Dimensions"] = [rows, cols]
    
    image_directories.append(img_aug_dest)

    for j in range(2, 41):
        if df["Object " + str(j - 1)][i] == 0:
            aug_dict["Object " + str(j - 1)] = 0
            continue
        
        bbox = copy.deepcopy(df["Object " + str(j - 1)][i])
        bbox[2] = int(rows) - int(bbox[2])
        bbox[4] = int(rows) - int(bbox[4])
        
        aug_dict["Object " + str(j - 1)] = bbox
    
    df = df.append(aug_dict, ignore_index=True)
    img_vert = cv2.flip(img_horiz, 0)
    
    if is_safe:
        plt.imsave(img_aug_dest, img_vert)
    

    # Rotation
    aug_dict = {}
    
    img_aug_name = df["image"][i][:-4] + "c" + ".png"
    img_aug_dest = 'archive/augmented_data/images/' + img_aug_name
    aug_dict["image"] = img_aug_name
    
    image_directories.append(img_aug_dest)
    
    angle = random.choice([90,180,270])
    if angle == 270:
        img_angle = np.transpose(img, (1,0,2))
        img_angle = cv2.flip(img_angle, 0)
    elif angle == 180:
        img_angle = cv2.flip(img_angle, -1)
    elif angle == 90:
        img_angle = np.transpose(img, (1,0,2))
        img_angle = cv2.flip(img_angle, 1)
    
    aug_dict["Dimensions"] = [img.shape[0], img.shape[1]]
    
    for j in range(2, 41):
        if df["Object " + str(j - 1)][i] == 0:
            aug_dict["Object " + str(j - 1)] = 0
            continue
        
        bbox = copy.deepcopy(df["Object " + str(j - 1)][i])
        
        x1 = int(bbox[1])
        x2 = int(bbox[3])
        y1 = int(bbox[2])
        y2 = int(bbox[4])
        if angle == 270:
            bbox[1] = y1
            bbox[3] = y2
            bbox[2] = cols - x2
            bbox[4] = cols - x1
        elif angle == 180:
            bbox[3] = cols - x1
            bbox[1] = cols - x2
            bbox[4] = rows - y1
            bbox[2] = rows - y2
        elif angle == 90:
            bbox[1] = rows - y2
            bbox[3] = rows - y1
            bbox[2] = x1
            bbox[4] = x2        
        
        aug_dict["Object " + str(j - 1)] = bbox
    
    df = df.append(aug_dict, ignore_index=True)
    
    if is_safe:
        plt.imsave(img_aug_dest, img_angle)
 
```

```python
from sklearn.decomposition import PCA
import numpy as np

img = img_arr[3]
reshaped_img = np.reshape(img, (img.shape[0], img.shape[1]*img.shape[2]))

v = []
for i in range(1, 40, 3):
    sum = 0
    for img in img_arr:
        img = img_arr[i]
        reshaped_img = np.reshape(img, (img.shape[0], img.shape[1]*img.shape[2]))
        
        pca = PCA(i).fit(reshaped_img) 
        img_transformed = pca.transform(reshaped_img)
        sum += np.sum(pca.explained_variance_ratio_)
    v.append(sum/len(img_arr))
temp = pca.inverse_transform(img_transformed) 
print(temp.shape)
temp = np.reshape(temp, (600,600 ,4)) 
print(temp.shape) 
plt.imshow(temp)
```
